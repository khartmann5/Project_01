```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = pd.read_csv("Resources/city_pop_transit_age_latlng.csv")
city_data.head()
```

```python
pub_trans = city_data.sort_values("Public Transportation Rate")
pub_trans.head()
```

```python
pub_trans = pub_trans.loc[pub_trans["Public Transportation Rate"] > 1]
pub_trans.count()
```

```python
pub_trans = pub_trans.sort_values("Population")
pub_trans.head()
```

```python
pub_trans = pub_trans.loc[pub_trans["Population"] >= 50000]
pub_trans = pub_trans.sort_values("Population")
pub_trans.head(20)
```

```python
small_city = pub_trans.loc[pub_trans["Population"] <= 60000]
small_city.head(20)
```

```python
small_city.to_csv("Resources/small_city.csv", index=False)
```
