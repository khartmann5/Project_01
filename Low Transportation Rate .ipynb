```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city_data = pd.read_csv("Resources/city_pop_transit_age_latlng.csv")
city_data.head()
```

```python
low_transportation = city_data.loc[city_data["Public Transportation Rate"] > .5]
low_transportation = low_transportation.sort_values("Public Transportation Rate")
low_transportation.head(20)
```

```python
low_transportation = low_transportation.loc[low_transportation["Population"] >= 50000]
low_transportation.head(20)
```

```python
low_trans_rate = low_transportation.loc[low_transportation["Public Transportation Rate"] < 1]
low_trans_rate.head(30)
```

```python
low_trans_rate.to_csv("Resources/low_rate.csv", index=False)
```
